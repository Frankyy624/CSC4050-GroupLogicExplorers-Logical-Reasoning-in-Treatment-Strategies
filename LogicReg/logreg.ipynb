```python
def logreg_anneal_control(start=0, end=0, iter=0, earlyout=0, update=0):
    if isinstance(start, dict):
        if 'end' in start and len(start['end']) > 0:
            end = start['end']
        if 'iter' in start and len(start['iter']) > 0:
            iter = start['iter']
        if 'earlyout' in start and len(start['earlyout']) > 0:
            earlyout = start['earlyout']
        if 'update' in start and len(start['update']) > 0:
            update = start['update']
        if 'start' in start and len(start['start']) > 0:
            start = start['start']
        else:
            start = 0
    if iter < 100 and (start != 0 or end != 0):
        raise ValueError("not enough repetitions")
    if start < end:
        raise ValueError("starting temperature below ending temperature")
    return {'start': start, 'end': end, 'iter': iter, 'earlyout': earlyout, 'update': update}

logreg_anneal_control(start=0, end=0, iter=0, earlyout=0, update=0)
```

```python
import numpy as np

def logreg_tree_control(treesize=8, opers=1, minmass=0, n1=None):
    if isinstance(treesize, dict):
        if 'opers' in treesize and len(treesize['opers']) > 0:
            opers = treesize['opers']
        if 'minmass' in treesize and len(treesize['minmass']) > 0:
            minmass = treesize['minmass']
        if 'treesize' in treesize and len(treesize['treesize']) > 0:
            treesize = treesize['treesize']
        else:
            treesize = 16
    
    treesize = 2 ** np.floor(np.log2(np.abs(treesize + 0.0001)))
    
    if opers != 2 and opers != 3 and opers != 4:
        opers = 1
    
    if minmass < 0:
        minmass = 0
    
    if n1 is not None:
        if minmass > n1 / 4:
            raise ValueError("minmass should be at most samplesize/4")
    
    return {'treesize': treesize, 'opers': opers, 'minmass': minmass}
```

```python

import inspect #for returning function parameters

def logreg(resp=None, bin=None, sep=None, wgt=None, cens=None, type_=None, select=None, 
           ntrees=None, nleaves=None, penalty=None,seed=None, kfold=None, nrep=None, 
           oldfit=None, anneal_control=None, tree_control=None, mc_control=None):
       
       args = inspect.getargvalues(inspect.currentframe())
       call = args.locals

       def logreg_binary(x):
              l1 = x.shape[0]*x.shape[1]
              l2 = np.sum(x == 0)
              l3 = np.sum(x == 1)
              return l1 == (l2 + l3)

       def logreg_storetree(x): #input: numpy array
              i1 = np.array(x[3:], dtype=float).reshape(-1, 4)
              i2 = i1.shape[0]
              i3 = pd.DataFrame({'number': np.arange(1, i2 + 1),
                                   'conc': i1[:, 0],
                                   'knot': i1[:, 1],
                                   'neg': i1[:, 2],
                                   'pick': i1[:, 3]})
              return i3

       n1 = 0
       miss_select = 1 if select is None else 0
       miss_sep = 1 if sep is None else 0
       miss_wgt = 1 if wgt is None else 0
       miss_cens = 1 if cens is None else 0
       miss_type = 1 if type is None else 0
       miss_ntrees = 1 if ntrees is None else 0
       miss_nleaves = 1 if nleaves is None else 0
       miss_penalty = 1 if penalty is None else 0
       miss_seed = 1 if seed is None else 0
       miss_kfold = 1 if kfold is None else 0
       miss_nrep = 1 if nrep is None else 0
       miss_tree = 1 if tree_control is None else 0
       miss_anneal = 1 if anneal_control is None else 0
       miss_mc = 1 if mc_control is None else 0
       miss_binnames = 1

       if oldfit is not None:
              if oldfit.class_ != 'logreg': 
                     raise ValueError("oldfit is not of class logreg")
              if resp is None and len(oldfit['response']) > 0:
                     resp = oldfit.response
              if bin is None and len(oldfit['binnames']) > 0:
                     bin = oldfit.binary
                     miss_binnames = 0
              if sep is None and oldfit['nsep'] > 0:
                     miss_sep = 0
                     sep = oldfit['separate']

              if cens is None and len(oldfit['censor']) > 0:
                     miss_cens = 0
                     cens = oldfit['censor']

              if wgt is None and len(oldfit['weight']) > 0:
                     miss_wgt = 0
                     wgt = oldfit['weight']

              if nrep is None and len(oldfit['nrep']) > 0:
                     miss_nrep = 0
                     nrep = oldfit['nrep']

              if kfold is None and len(oldfit.kfold) > 0:
                     miss_kfold = 0
                     kfold = oldfit['kfold']

              if penalty is None and len(oldfit.penalty) > 0:
                     miss_penalty = 0
                     penalty = oldfit['penalty']

              if anneal_control is None and len(oldfit.anneal_control) > 0:
                     miss_anneal = 0
                     anneal_control = oldfit['anneal_control']

              if mc_control is None and len(oldfit['mc_control']) > 0:
                     miss_mc = 0
                     mc_control = oldfit['mc_control']

              if tree_control is None and len(oldfit['tree_control']) > 0:
                     miss_tree = 0
                     tree_control = oldfit.tree_control

              if ntrees is None and len(oldfit['ntrees']) > 0:
                     miss_ntrees = 0
                     ntrees = oldfit.ntrees
                     if len(ntrees) == 1:
                            ntrees = [ntrees[0], ntrees[0]]

              if nleaves is None and len(oldfit['nleaves']) > 0:
                     miss_nleaves = 0
                     nleaves = oldfit['nleaves']
                     if len(nleaves) == 1:
                            nleaves = [nleaves[0], nleaves[0]]

              if seed is None and len(oldfit['seed']) > 0:
                     miss_seed = 0
                     seed = oldfit.seed

              if type_ is None and len(oldfit['type']) > 0:
                     miss_type = 0
                     if oldfit['type'] == "own.scoring":
                            type_ = 0
                     if oldfit['type'] == "classification":
                            type_ = 1
                     if oldfit['type'] == "regression":
                            type_ = 2
                     if oldfit['type'] == "logistic":
                            type_ = 3
                     if oldfit['type'] == "proportional.hazards":
                            type_ = 4
                     if oldfit['type'] == "exponential.survival":
                            type_ = 5

              if select is None and len(oldfit['select']) > 0:
                     miss_select = 0
                     if oldfit['select'] == "single.model":
                            select = 1
                     if oldfit['select'] == "multiple.models":
                            select = 2
                     if oldfit['select'] == "cross.validation":
                            select = 3
                     if oldfit['select'] == "null.model.test":
                            select = 4
                     if oldfit['select'] == "randomization.test":
                            select = 5
                     if oldfit['select'] == "bayesian":
                            select = 7
                     if oldfit['select'] == "greedy":
                            select = 6
       if miss_type == 1:
              type_ = 2
       if miss_penalty == 1:
              penalty = 0
       if miss_kfold == 1:
              kfold = 10
       if miss_nrep == 1:
              nrep = 25
       if miss_seed == 1:
              seed = 0
       if miss_binnames == 1:
              if isinstance(bin, pd.DataFrame):
                     binnames = list(bin.columns)
              else:
                     binnames = None
       if isinstance(type_, int):
              mdl = type_
       else:
              type_lower = type.lower()
              if type_lower == "c":
                     mdl = 1
              elif type_lower == "r":
                     mdl = 2
              elif type_lower == "l":
                     mdl = 3
              elif type_lower == "s":
                     mdl = 4
              elif type_lower == "e":
                     mdl = 5
              elif type_lower == "o":
                     mdl = 0

       if mdl < 0 or mdl > 5:
              raise ValueError("not an implemented model")
       choice = 99
       if select.isdigit():  
              choice = int(select)
       else:
              select = select.lower()  
              if select == "s":
                     choice = 1
              elif select == "m":
                     choice = 2
              elif select == "c":
                     choice = 3
              elif select == "n":
                     choice = 4
              elif select == "r":
                     choice = 5
              elif select == "b":
                     choice = 7
              elif select == "g":
                     choice = 6
       if choice < 0 or choice > 7:
              raise ValueError("not a valid selection approach")

       n1 = len(resp)
       if (mdl in [1, 3]) and not logreg_binary(resp):
              raise ValueError("some non binary response data, and a binary fitting function")
       if (mdl in [4, 5]) and (min(resp) < 0):
              raise ValueError("survival data needs to be positive")

       bin = np.array(bin)
       n2 = len(bin) / n1
       if n2 != np.floor(n2):
              raise ValueError("length of binary data not a multiple of response")
       if n2 < 1:
              raise ValueError("binary predictors mandatory")
       if not logreg_binary(bin):
              raise ValueError("some non binary data among binary predictors")

       nsep = 0
       if miss_sep == 0:
              sep = np.array(sep)
              nsep = len(sep) / n1
       if nsep != np.floor(nsep):
              raise ValueError("length of separate predictors not a multiple of response")
       else:
              sep = 0

       if nsep > 0 and (mdl == 2 or mdl == 5 or mdl == 4):
              tep = sep.copy()
              k = 0
              for i in range(nsep):
                     if (np.min(tep[:, i]) != 0 or np.max(tep[:, i]) != 1) and k != 2:
                            k = 1
                            if np.max(tep[:, i]) != np.min(tep[:, i]):
                                   tep[:, i] = (tep[:, i] - np.min(tep[:, i])) / (np.max(tep[:, i]) - np.min(tep[:, i]))
                                   if not logreg_binary(tep[:, i]):
                                          k = 2
                            else:
                                   tep[:, i] = 0

              if k == 1 and mdl != 4:
                     sep = tep
                     print("separate covariates that were binary recoded to 0/1")

              if k == 2 and mdl != 4:
                     if mdl == 5:
                            raise ValueError("Exponential survival models are only implemented if\nall separate predictors are binary")
                     if mdl == 3:
                            print("Logistic regression runs much faster if\nall separate predictors are binary")

              if mdl == 4 and k != 2:
                     print("The code will be run much faster if you use exponential regression\n"
                            "instead of proportional hazards models after a cumulative hazards\n"
                            "transformation, as all your separate predictors are binary\n"
                            "type help(cumhaz) for more information\n")

       if miss_tree == 1:
              tree_control = logreg_tree_control(n1=n1)
       else:
              tree_control = logreg_tree_control(tree_control, n1=n1)

       if miss_wgt == 1:
              wgt = [1] * n1
       else:
              if len(wgt) != n1:
                     raise ValueError("weight does not have the right length")
              if min(wgt) < 0 and mdl != 0:
                     raise ValueError("weight has to be larger than 0")

       if miss_cens == 1:
              cens = [1] * n1
       else:
              if len(cens) != n1:
                     raise ValueError("cens doesn't have right length")
              if (mdl == 4 or mdl == 5) and (not logreg_binary(cens)):
                     raise ValueError("censoring data has to be 0/1")

       if choice == 1 or choice == 4 or choice == 7:
              if miss_ntrees:
                     ntr = [1, 1]
              else:
                     ntr = [ntrees[0]] * 2
              if miss_nleaves:
                     msz = [-1, -1]
              else:
                     msz = [nleaves[0]] * 2
       else: 
              if miss_ntrees:
                     ntr = [1, 1]
              else:
                     ntr = ntrees[0:2]
              if np.isnan(ntr[1]):
                     ntr[1] = ntr[0]
                     if choice == 6 and miss_nleaves:
                            nleaves = [tree_control['treesize'] * ntr[1]] * 2

              msz = nleaves[0:2]
              if np.isnan(msz[1]):
                     msz[1] = msz[0]

              if ntr[0] > ntr[1]:
                     raise ValueError("upper limit of number of trees smaller than lower limit")
              if msz[0] > msz[1]:
                     raise ValueError("upper limit of number of leaves smaller than lower limit")
              if msz[0] < 0:
                     raise ValueError("number of leaves needs to be at least 0")

       if ntr[0] < 1:
              raise ValueError("number of trees needs to be at least 1")
       if mdl == 1 and max(ntr) > 1:
              raise ValueError("for classification only 1 tree is possible")
       if mdl == 1 and nsep > 0:
              raise ValueError("for classification no separate predictors are possible")
       if miss_anneal == 1:
              anneal_control = logreg_anneal_control()
       else:
              anneal_control = logreg_anneal_control(anneal_control)
       if miss_mc == 1:
              mc_control = logreg_mc_control()
       else:
              mc_control = logreg_mc_control(mc_control)
       if miss_mc != 0 and select == 7:
              anneal_control['update'] = mc_control['update']
       if penalty < 0 and (choice == 1 or choice == 6):
              raise ValueError("penalty should be at least 0")
       if choice == 3:
              nrep = kfold
              if nrep < 2:
                     raise ValueError("kfold needs to be at least 2")
              
       if choice > 3 and nrep < 1 and choice < 6:
              raise ValueError("nrep needs to be at least 1")
       
       xseed = seed
       if xseed == 0:
              xseed = int(np.floor(np.random.uniform(1) * 1000000) + 1)

       ipars = [mdl, msz[0], msz[1], tree_control['treesize'], ntr[0], ntr[1], tree_control['opers'],
              anneal_control['update'], xseed, nrep, choice, anneal_control['earlyout'],
              tree_control['minmass'], mc_control['nburn'], mc_control['niter'], mc_control['output']]
       rpars = [anneal_control['start'], anneal_control['end'], anneal_control['iter'], penalty,
              mc_control['hyperpars']]
       if len(rpars) < 14:
              rpars += [0] * (14 - len(rpars))
       orders = np.argsort(np.argsort(resp) + np.random.uniform(size=n1) / 1000000)
       nkn = tree_control['treesize'] * 2 - 1
       nxx = 2

       if choice == 1 or choice == 7:
              na = ntr[0] * (nkn * 4 + 3)
              nb = nsep + ntr[0] + 1
              nc = 2
       elif choice == 2:
              nd = (ntr[1] - ntr[0] + 1) * (msz[1] - msz[0] + 1)
              na = nd * ntr[1] * (nkn * 4 + 3)
              nb = nd * (nsep + ntr[1] + 1)
              nc = nd
       elif choice == 6:
              nd = msz[1] + 2
              na = nd * ntr[1] * (nkn * 4 + 3)
              nb = nd * (nsep + ntr[1] + 1)
              nc = nd
       elif choice == 3:
              na = nb = 2
              nd = (ntr[1] - ntr[0] + 1) * (msz[1] - msz[0] + 1)
              nc = nd * 8 * nrep
       elif choice == 4:
              na = nb = 2
              nc = nrep + 2
       n100 = -100
       if choice == 7:
              na = 256
              nb = n2
              nc = n2 * n2
              if abs(mc_control['output']) < 2:
                     nc = 1
              nxx = nc * n2
              if abs(mc_control['output']) < 3:
                     nxx = 1
              n100 = 0
       if choice != 5:
              xtree = [n100] * na
       if choice == 5:
              nb = 2
              nd = (ntr[1] - ntr[0] + 1) * (msz[1] - msz[0] + 1)
              nc = (nrep + 2) * nd + 2
              xtree = None
              if tree_control['treesize'] != oldfit['tree_control']['treesize']:
                     raise ValueError("treesize should match treesize in oldfit")
              if oldfit['choice'] != 2:
                     raise ValueError("oldfit not an object from a multiple model fit")
              if oldfit['choice'] == 2:
                     for i in range(1, oldfit['nmodels'] + 1):
                            tmp = oldfit['alltrees'][i - 1]
                            if i == 1:
                                   bestscore = tmp['score']
                            else:
                                   xscore = tmp['score']
                            if xscore < bestscore:
                                   bestscore = xscore
                            for j in range(tmp['ntrees'][0]):
                                   xtree.append(tmp['nleaves'][0])
                                   xtree.append(tmp['ntrees'][0])
                                   xtree.append(j)
                                   for tree_row in np.array(tmp['trees'][j - 1]['trees'])[:, 1:]:
                                          xtree.extend(tree_row)
                     allscores = oldfit['allscores']
              xtree.extend([n100] * (4 * nkn * 10))
       ip4 = 2 * ipars[3] + 1

       ###########  c function   ################################################
       fit = {'n1': int(n1),
              'n2': int(n2),
              'nsep': int(nsep),
              'ip': np.array(ipars, dtype=np.int),
              'rpars': np.array(rpars, dtype=np.int),
              'sep' : np.array(sep, dtype=np.single).T,
              'cens': np.array(cens, dtype=int),
              'orders' : np.array(orders, dtype=int),
              'resp' : np.array(resp, dtype=np.single),
              'wgt': np.array(wgt, dtype=np.single),
              'bin': np.array(bin, dtype=np.int).T,
              'trees': np.array(xtree, dtype=np.int),
              'coef' : np.full(nb, n100, dtype=np.single),
              'scores': np.full(nc, n100, dtype=np.single),
              'rd4' : np.zeros(nxx, dtype=np.int32),
              }
       
       if fit['ip'][0] < -900:
              raise ValueError("fatal declaration error - reduce problem or recompile package")

       ipars = choice
       rpars = [penalty]
       # Map mdl values to type strings
       mdl_map = {
                     0: "own.scoring",
                     1: "classification",
                     2: "regression",
                     3: "logistic",
                     4: "proportional.hazards",
                     5: "exponential.survival"
                     }

       # Map choice values to chs strings
       choice_map = {
                     1: "single.model",
                     2: "multiple.models",
                     3: "cross.validation",
                     4: "null.model.test",
                     5: "randomization.test",
                     7: "bayesian",
                     6: "greedy"
                     }

       # Assign type and chs based on mdl and choice values
       type = mdl_map.get(mdl, "unknown")
       chs = choice_map.get(choice, "unknown")

       if tree_control['opers'] != 2 and tree_control['opers'] != 3:
              tree_control['operators'] = "both"
       elif tree_control['opers'] == 2:
              tree_control['operators'] = "and"
       else:
              tree_control['operators'] = "or"

       m1 = {
              "nsample": n1,
              "nbinary": n2,
              "nseparate": nsep,
              "type": type,
              "select": chs,
              "anneal_control": anneal_control,
              "tree_control": tree_control,
              "seed": seed,
              "choice": choice
              }
       
       if choice == 7:
              m1['anneal_control'] = None
              v1 = fit['trees']
              v3 = max([i for i, val in enumerate(v1) if val > 0])
              v1 = v1[:v3]
              v1 = pd.DataFrame({'size': range(v3), 'number': v1})
              m1['size'] = v1
              m1['single'] = fit['coef']
              m1['single'][m1['single'] < 0] = 0
              m1['mc_control'] = mc_control
              if abs(mc_control['output']) > 1:
                     m1['double'] = np.where(fit['scores'] < 0, 0, fit['scores'])
              if abs(mc_control['output']) > 2:
                     m1['triple'] = np.array(fit['rd4']).reshape(len(fit['coef']), len(fit['coef']), len(fit['coef']))


       if choice == 1 or choice == 4:
              m1['nleaves'] = msz[0]
              m1['ntrees'] = ntr[0]
       else:
              m1['nleaves'] = msz
              m1['ntrees'] = ntr

       if choice == 4 or choice == 5:
              m1['nrep'] = nrep
       if choice == 3:
              m1['kfold'] = kfold
       if choice == 1:
              m1['penalty'] = penalty

       m1['response'] = resp
       m1['binary'] = bin
       m1['separate'] = sep
       m1['censor'] = cens
       m1['weight'] = wgt

       if choice == 3:
              cvscores = np.array(fit['scores']).reshape(-1, 8) #cross validation scores
              cvscores = cvscores[cvscores[:, 0] > -99] #筛出第一列大于-99的行
              cvscores = {
                     "ntree": cvscores[:, 0],
                     "nleaf": cvscores[:, 1],
                     "k": cvscores[:, 2],
                     "kfold": cvscores[:, 3],
                     "train": cvscores[:, 4],
                     "train.ave": cvscores[:, 5],
                     "test": cvscores[:, 6],
                     "test.ave": cvscores[:, 7]
              }
              m1['cvscores'] = cvscores

       if choice == 4:
              m1['nullscore'] = fit['scores'][0]
              m1['bestscore'] = fit['scores'][1]
              m1['randscore'] = fit['scores'][2:(2 + nrep)]

       if choice == 5:
              m1['nullscore'] = fit['scores'][0]
              m1['bestscore'] = fit['scores'][1]
              m1['allscores'] = allscores
              randscores = np.array(fit['scores'][2:])
              randscores = randscores.reshape(nrep + 2, -1)
              randscores = randscores[:, np.all(randscores[0] > -99, axis=0)]

              randscores_dict = {"ntree": randscores[0], "nleaf": randscores[1]}
              for i, score in enumerate(randscores[2:], start=1):
                     randscores_dict[f"{i}"] = score

              m1['randscores'] = randscores_dict


       if choice == 1:
              m2 = {}
              m2['ntrees'] = ntr
              m2['nleaves'] = msz
              m2['score'] = fit['scores'][0]
              m2['coef'] = fit['coef'][:nsep + ntr[0] + 1]
              if mdl == 4:
                     m2['coef'][0] = 0
              if mdl == 1:
                     m2['coef'][:2] = [0, 1]
              m2['class'] = "logregmodel"
              lx = 3 + 4 * nkn
              m2['trees'] = []
              for i in range(ntr[0]):
                     m3 = {} # an element in 'trees':[{...},...,{...}]
                     m3['coef'] = fit['coef'][1 + nsep + i]
                     if mdl == 1:
                            m3['coef'] = None
                            ######
                            m3['trees'] = logreg_storetree(fit['trees'][(i - 1) * lx + np.arange(1, lx + 1)]) 
                            ######
                            m3['class'] = "logregtree"
                            m2['trees'].append(m3)

              m1['model'] = m2

       if choice == 6:
              m1['nmodels'] = fit['ip'][0]
              m1['alltrees'] = []
              m1['allscores'] = np.zeros((m1['nmodels'], 3))
              j = 0
              k = 0
              for i in range(1, m1['nmodels'] + 1):
                     m2 = {}
                     m2['score'] = fit['scores'][i - 1]
                     m2['nleaves'] = fit['trees'][j]
                     m2['ntrees'] = fit['trees'][j + 1]
                     m1['allscores'][i - 1] = np.array([fit['scores'][i - 1], fit['trees'][j], fit['trees'][j + 1]])
                     m2['coef'] = fit['coef'][k:k + m2['ntrees'] + nsep + 1]
                     if mdl == 4:
                            m2['coef'][0] = 0
                     if mdl == 1:
                            m2['coef'][:2] = [0, 1]

                     m2['class'] = "logregmodel"
                     m2['trees'] = []
                     lx = 3 + 4 * nkn
                     for l in range(m2['ntrees']):
                            m3 = {}
                            m3['whichtree'] = l
                            m3['coef'] = m2['coef'][1 + nsep + l]
                            if mdl == 1:
                                   m3['coef'] = None
                            m3['trees'] = logreg.storetree(fit['trees'][j + (l - 1) * lx:j + (l * lx)])
                            m3['class'] = "logregtree"
                            m2['trees'].append(m3)

                     j += lx * m2['ntrees']
                     k += m2['ntrees'] + nsep + 1
                     m1['alltrees'].append(m2)

       if choice == 2:
              m1['nmodels'] = fit['ip'][0]
              m1['alltrees'] = []
              m1['allscores'] = np.zeros((m1['nmodels'], 3))
              j = 0
              k = 0
              for i in range(1, m1['nmodels'] + 1):
                     m2 = {}
                     m2['score'] = fit['scores'][i - 1]
                     m2['nleaves'] = fit['trees'][j]
                     m2['ntrees'] = fit['trees'][j + 1]
                     m1['allscores'][i - 1, :] = [fit['scores'][i - 1], fit['trees'][j], fit['trees'][j + 1]]
                     m2['coef'] = fit['coef'][k:k + m2['ntrees'] + nsep + 1]
                     if mdl == 4:
                            m2['coef'][0] = 0
                     if mdl == 1:
                            m2['coef'][:2] = [0, 1]

                     m2['class'] = 'logregmodel'
                     m2['trees'] = []
                     lx = 3 + 4 * nkn
                     for l in range(m2['ntrees']):
                            m3 = {}
                            m3['whichtree'] = l
                            m3['coef'] = m2['coef'][1 + nsep + l]
                            if mdl == 1:
                                   m3['coef'] = None
                            m3['trees'] = logreg.storetree(fit['trees'][j + (l - 1) * lx : j + l * lx])
                            m3['class'] = 'logregtree'
                            m2['trees'].append(m3)
                     j += lx * m2['ntrees']
                     k += m2['ntrees'] + nsep + 1
                     m1['alltrees'].append(m2)
       #call
       m1['call'] = call
       m1['binnames'] = binnames
       m1['class'] = 'logreg'
       
       return m1








       
       
```
